```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

```python
#SCATTER PLOT
df = pd.DataFrame({
    'x': np.random.rand(100),
    'y': np.random.rand(100),
    'category': np.random.choice(['A','B','C'], 100)
})

plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x='x', y='y', hue='category', style='category', s=100, palette='deep')
plt.title('Seaborn Scatter Plot')
plt.show()
```

```python
#Line Plot
x = np.linspace(0,10,100)
y = np.sin(x)

plt.figure(figsize=(16,6))
sns.lineplot(x=x, y=y, color='blue', linewidth=2)
plt.title('Seaborn Line Plot')
plt.show()
```

```python
#Histogram
data = np.random.randn(100)
plt.figure(figsize=(16,6))
sns.histplot(data, bins=30, kde=True, color='green')
plt.title('Histogram Graph')
plt.show()
```

```python
#Kernel Density Estimate (KDE) (sns.kdeplot)
data = np.random.randn(100)
plt.figure(figsize=(16,6))
sns.kdeplot(data, shade=True, color='blue')
plt.title('Kernel Density Estimate(KDE)')
plt.show()
```

```python
#Bar Plot
df = pd.DataFrame({
    'category':['A','B','C','D'],
    'values':[10,20,30,40]
})
plt.figure(figsize=(16,6))
sns.barplot(data=df, x='category', y='values', palette='viridis')
plt.title('Bar Plot')
plt.show()
```

```python
#COUNT PLOT
df = pd.DataFrame({'category':np.random.choice(['A','B','C'],100)})
plt.figure(figsize=(16,6))
sns.countplot(data=df, x='category', palette='pastel')
plt.title('Count Plot')
plt.show()
```

```python
#Swarm Plot
df = pd.DataFrame({
    'category': np.random.choice(['A','B','C'], 100),
    'values': np.random.randn(100)
})
plt.figure(figsize=(16,6))
sns.swarmplot(data=df, x='category',y='values',palette='Set2')
plt.title('Swarm Plot')
plt.show()
```

```python
#Violin Plot
df = pd.DataFrame({
    'category': np.random.choice(['A','B','C'], 100),
    'values': np.random.randn(100)
})
plt.figure(figsize=(8,5))
sns.violinplot(data=df, x='category', y='values', palette='Set3', inner='quartile')
plt.title("Seaborn Violin Plot")
plt.show()
```
