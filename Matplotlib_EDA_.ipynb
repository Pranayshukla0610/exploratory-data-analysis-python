```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)

plt.figure(figsize=(16,3))
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.title("Basic Line Plot")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Scatter Plot
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(16,6))
plt.scatter(x,y, color='red', marker='o', s=100, alpha=0.7)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
#Bar Plot
categories = ['A','B','C','D']
values = [10,15,20,25]

plt.figure(figsize=(16,6))
plt.bar(categories, values, color='green')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
#Horizontal Bar Chart
categories = ['A','B','C','D','E','F','G','H']
values = [10,5,25,7,50,45,43,55]
plt.figure(figsize=(16,6))
plt.barh(categories, values, color='red')
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
#Histogram
data = np.random.randn(1000)
plt.figure(figsize=(16,6))
plt.hist(data, bins=30, color='blue', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Frequency')
plt.ylabel('Values')
plt.show()
```

```python
#Pie Chart
sizes = [15, 30, 25, 14]
fruits = ['Apple','WaterMelon','Papaya','Banana']

plt.figure(figsize=(16,6))
plt.pie(sizes,labels=fruits, autopct='%1.1f%%', startangle=140, colors=['red','green','orange','yellow'])
plt.title('Pie Chart')
plt.show()
```

```python
#ADVANCED 2D PLOTS
#Multiple Lines on Same Plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(16,6))
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='-', linewidth=2)
plt.title('Multiple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Stack Plot
days = [1,2,3,4,5]
sleeping=[5,6,7,8,9]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]

plt.figure(figsize=(16,6))
plt.stackplot(days, sleeping, eating, working, labels=['Sleeping','Eating','Working'], colors=['m','c','r'])
plt.title('Stack Plot')
plt.xlabel('Days')
plt.ylabel('Hours')
plt.legend(loc='upper left')
plt.show()
```

```python
#Area Plot
x = np.arange(1, 6)
y1 = np.array([1,3,5,7,9])
y2 = np.array([2,4,6,8,10])

plt.figure(figsize=(16,6))
plt.fill_between(x,y1,y2,color='skyblue',alpha=0.4)
plt.plot(x,y1,color='blue',alpha=0.7,linewidth=2)
plt.plot(x,y2,color='magenta',alpha=0.7,linewidth=2)
plt.title('Area Plot')
plt.show()
```

```python
#Customizing Styles
plt.style.use('ggplot')
x = np.linspace(0,10,100)
y = np.sin(x)

plt.figure(figsize=(16,6))
plt.plot(x, y, label='sin(x)', color='darkblue', linestyle='-', linewidth=3, marker='o',markersize=5)
plt.title('ggplot')
plt.show()
```
