```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
sns.pairplot(df, hue='species', palette='bright')
plt.suptitle('Seaborn Pair Plot',y=1.02)
plt.show()
```

```python
#Joint Plot
df = sns.load_dataset('iris')
sns.jointplot(x='sepal_length', y='sepal_width', kind='scatter', hue='species', data=df, palette='bright')
plt.show()
```

```python
#Hexbin / Hex Plot (sns.jointplot with kind='hex')
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'x': np.random.randn(1000),
    'y': np.random.randn(1000)
})
sns.jointplot(data=df, x='x', y='y', kind='hex', color='purple')
plt.show()
```

```python
#Heatmap (sns.heatmap)
data = np.random.rand(5,5)
plt.figure(figsize=(16,6))
sns.heatmap(data, annot=True, cmap='Blues', linewidths=0.5)
plt.title('Heatmap')
plt.show()
```

```python
#Clustermap (sns.clustermap)
df = np.random.rand(16,6)
sns.clustermap(df, cmap='coolwarm', annot=True, standard_scale=1)
plt.title('Clustermap',y=1.05)
plt.show()
```

```python
#Regression Plot (sns.regplot)
df = sns.load_dataset('tips')
plt.figure(figsize=(8,5))
sns.regplot(data=df, x='total_bill', y='tip', scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title("Seaborn Regression Plot")
plt.show()
```

```python
#Residual Plot (sns.residplot)
plt.figure(figsize=(8,5))
sns.residplot(data=df, x='total_bill', y='tip', color='green')
plt.title("Seaborn Residual Plot")
plt.show()
```

```python
#Facet Grid (sns.FacetGrid)
g = sns.FacetGrid(df, col="sex", row="time", margin_titles=True)
g.map(sns.scatterplot, "total_bill", "tip", alpha=0.7)
g.add_legend()
plt.show()
```
